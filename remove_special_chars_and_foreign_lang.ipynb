```python
import nltk
import csv
import itertools
import re
```

```python
words = set(nltk.corpus.words.words())
```

```python
link='condensedStats.csv'

with open(link, newline='', encoding='utf8') as f:
    lines = itertools.islice(f,1,None)
    reader = csv.reader(lines)
    
    # Header
    list1=['Id', 'Title', 'Description', 'LikeCount', 'DislikeCount', 'Location (latitude, longitude)', 'Condensed Tags (comma delimited string)']
    final_list=[]
    final_list.append(list1)
    for row in reader:
        # Remove non English words
        check1=" ".join(w for w in nltk.wordpunct_tokenize(row[1]) \
                if w.lower() in words or not w.isalpha())

        check2=" ".join(w for w in nltk.wordpunct_tokenize(row[2]) \
                if w.lower() in words or not w.isalpha())


        check3=" ".join(w for w in nltk.wordpunct_tokenize(row[6]) \
                if w.lower() in words or not w.isalpha())

        flag=0
        # If title or description or tags are empty
        if(len(row[1])==0 or len(row[2])==0 or len(row[6])==0):
            flag=1
        # Threshold value is set to 0.5
        # If length of string after removing non English words from 
        # title or description or tags is below the threshold then the video is not included in the data set
        elif(((len(check1)/len(row[1]))<0.5) or ((len(check2)/len(row[2]))<0.5) or ((len(check3)/len(row[6]))<0.5)):
            flag=1
        if(flag==0):
            #writer = csv.writer(f)
            list1=[]
            for i in range(0,len(row)):
                if(i==6):
                    # Remove empty strings and special characters from tags
                    sent=row[6]
                    
                    check_arr=sent.split(",")

                    sent1=[]
                    for ele in check_arr:
                        if ele:
                            sent1.append(re.sub('[^A-Za-z0-9]+',' ',ele))

                    sent2=[]
                    for ele in sent1:
                        if len(ele)>1 or ele.isalnum():
                            sent2.append(ele.lstrip())
                    print(sent2)
                    row[6]=', '.join(sent2)         
                list1.append(row[i])
            final_list.append(list1)
    print(final_list)
```

```python
write_link='english_condensedStats.csv'

with open(write_link, 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerows(final_list)
```
